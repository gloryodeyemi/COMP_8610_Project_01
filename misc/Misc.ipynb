```python
def value_count_graph(data, var):
    data[var].value_counts().plot(kind="bar")
    plt.title("Value counts of the " + var + " variable")
    plt.xlabel(var)
    plt.xticks(rotation=0)
    plt.ylabel("Count")
    plt.show()
```

```python
value_count_graph(client_data, 'default_payment_next_month')
```

```python
client_data.default_payment_next_month.value_counts().plot(kind="bar")
plt.title("Value counts of the target variable")
plt.xlabel("Default payment: Yes=1, No=0")
plt.xticks(rotation=0)
plt.ylabel("Count")
plt.show()
```

```python
for var in x_list:
  sns.catplot(x="default_payment_next_month", y=var, data=client_data, kind="box", aspect=1.5)
  plt.title("Boxplot for default payment vs " + var)
  plt.show()
```

```python
rcParams['figure.figsize'] = (30, 15)
corr_mat = client_data.corr()
heat_map = sns.heatmap(corr_mat, 
                 cbar=True, 
                 annot=True, 
                 square=True, 
                 fmt='.2f', 
                 annot_kws={'size': 10}, 
                 yticklabels=client_data.columns, 
                 xticklabels=client_data.columns, 
                 cmap="Spectral_r")
plt.savefig('heatmap.png')
plt.show()
```
